```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

# fetch dataset 
automobile = fetch_ucirepo(id=10) 
automobile

data = automobile.data.features
```

```python
df_filt = data.loc[:,(-data.isnull().any())]
```

```python
df_filt
```

```python
df_filt.describe()
```

```python
df_filt.drop("engine-location", axis=1, inplace=True)
```

```python
df_filt.drop("wheel-base", axis=1, inplace=True)
```

```python
df_filt.drop("city-mpg", axis=1, inplace=True)
```

```python
df_filt.drop("num-of-cylinders", axis=1, inplace=True)
```

```python
df_filt
```

```python
#for storing into data base

import sqlite3
import pandas as pd

conn = sqlite3.connect("../database/automobile.db")
cur = conn.cursor()

df_filt.to_sql('auto_parts', conn, index=True, if_exists="replace") 
```

```python
rows = cur.execute("SELECT * FROM auto_parts").fetchall()
rows[0]
```
